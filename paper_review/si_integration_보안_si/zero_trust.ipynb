## ✅ Day 1 – Introduction & Motivation

---

### 📌 연구 배경과 동기
- **보안 경계의 붕괴 (Collapse of Perimeter Security)**  
  전통적인 보안 모델은 “내부 네트워크는 신뢰, 외부는 불신”이라는 가정을 전제로 했다.  
  그러나 클라우드·모바일·IoT 기기의 급격한 확산으로 인해,  
  네트워크의 경계가 모호해지고 내부에서도 공격이 발생하는 **‘경계 무의미화(De-Perimeterization)’** 현상이 나타났다.  

- **현대 인프라의 복잡성**  
  - 클라우드와 IoT가 결합된 환경에서는 수많은 엔드포인트와 API가 상호 연결되어 있으며,  
    단일 방화벽이나 VPN만으로는 전체 흐름을 보호할 수 없다.  
  - 공격자는 내부 계정 탈취, 공급망 침투(Supply Chain Attack) 등 **내부 침투형 공격**을 통해  
    기존 경계보안 체계를 우회한다.  

- **핵심 연구 질문**  
  “모든 접속 주체(사용자, 기기, 서비스)를 불신하고,  
  매 요청마다 검증하는 보안 구조를 어떻게 클라우드·IoT 환경에 통합할 수 있을까?”

<br>

### 📌 핵심 아이디어
- “**Never Trust, Always Verify**” 원칙에 기반한 **Zero Trust Architecture (ZTA)** 를 클라우드와 IoT 통합 구조에 적용한다.  
- 각 엔티티(사용자, 기기, 서비스)의 **동적 검증(Dynamic Verification)** 과  
  **정책 기반 접근 제어(Policy-based Access Control)** 를 수행한다.  
- 네트워크를 **Micro-Segmentation(미세 구획화)** 하여,  
  침입자가 하나의 구역을 뚫더라도 다른 영역으로의 이동(lateral movement)을 차단한다.  

<br>

### 📌 주요 기여
1. **ZTA의 IoT·Cloud 통합 구조 제안**  
   기존 Zero Trust 모델을 IoT와 클라우드가 혼재된 환경에 맞게 재설계하였다.  
2. **정책 중심 접근 제어 (Policy-Driven Control)**  
   사용자의 위치나 네트워크 신뢰도 대신, 정책 엔진이 **상황(Context)** 기반으로 접근 권한을 부여한다.  
3. **지속적 인증 (Continuous Authentication)**  
   단발성 인증이 아닌, 세션 중에도 지속적으로 신뢰 상태를 검증하여 위협에 실시간 대응한다.  
4. **보안·성능 균형 모델**  
   인증 빈도와 네트워크 오버헤드를 최소화하면서도 높은 보안성을 유지할 수 있는 구조를 설계하였다.  

<br>

### 📌 핵심 인사이트
- 보안은 더 이상 **“경계를 세우는 행위”가 아니라, “신뢰를 세분화하는 과정”** 이다.  
- Zero Trust는 기존의 “내부 신뢰” 개념을 폐기하고,  
  **“모든 요청은 잠재적으로 악성”** 이라는 가정에서 출발한다.  
- 이 논문은 보안관제(SOC)의 탐지 이후 대응 단계를 넘어,  
  **보안 SI 엔지니어링(사전 예방형 설계)** 으로 초점을 옮긴 대표 연구다.  

<br>

### 📌 핵심 문장 인용
> *“We propose a Zero Trust-based security framework that dynamically validates every entity and data flow in cloud and IoT systems, ensuring continuous verification and policy enforcement.”*  
> — *IEEE, 2021*

<br>

### 🧠 개인적 해석
- 이 논문은 “**신뢰를 없애는 것이 진정한 보안의 출발점**”이라는 철학을 제시한다.  
- SOC의 ‘탐지형 AI’에서 SI의 ‘설계형 보안’으로 넘어가는 전환점을 잘 보여준다.  
- Zero Trust는 단순히 기술 트렌드가 아니라, **보안 시스템 통합(SI) 패러다임의 변화**를 상징한다.  
- 앞으로 읽게 될 Day 2~5에서, 이 구조가 **어떻게 구현되고 평가되는지**를 집중적으로 따라가면 좋다.


## ✅ Day 2 – Related Work & Conceptual Framework

---

### 📌 기존 연구 및 한계 (Related Work)
- **Perimeter-based Security Model (경계보안 모델)**  
  - 네트워크 외부와 내부를 구분하고, 내부는 기본적으로 신뢰하는 구조.  
  - VPN, 방화벽, IDS/IPS 기반 접근 제어가 핵심.  
  - 하지만 클라우드·모바일 환경에서는 경계가 사라져 **‘내부=신뢰’ 가정이 무너짐.**

- **Identity-based Security (ID 중심 보안)**  
  - 사용자 인증(예: SSO, IAM)을 통해 접근 통제 수행.  
  - 계정 단위 제어에는 효과적이지만, **기기·데이터·세션 단위의 동적 검증이 불가능.**

- **Network Segmentation & Access Control Lists (ACL)**  
  - 트래픽을 영역별로 분리하고, ACL 정책으로 접근을 통제.  
  - 그러나 정책이 정적(Static)이라, 환경 변화나 위협 상황에 실시간 대응이 어렵다.

- **Cloud Access Security Broker (CASB)**  
  - SaaS·클라우드 서비스 간의 접근 제어를 중재하지만,  
    IoT 디바이스나 온프레미스 시스템까지 통합하기에는 한계가 있다.

> 💡 기존 모델들의 공통 한계:  
> 보안 정책이 **네트워크 단위 중심**, **정적 정책 기반**,  
> 그리고 **사용자·기기·데이터 간 관계를 통합적으로 고려하지 못함.**

---

### 📌 제안된 Zero Trust Framework 개요 (Conceptual Framework)

Zero Trust Architecture(ZTA)는 **“Policy-driven, Context-aware, Continuous Verification”** 세 가지 축으로 구성된다.  
이 논문에서는 클라우드와 IoT 환경을 아우르는 구조로 다음 세 핵심 컴포넌트를 제시한다.

---

#### 1️⃣ **Policy Engine (정책 엔진)**
- 모든 접근 요청을 중앙에서 판단하는 핵심 제어 모듈.  
- 사용자의 신원(ID), 기기의 무결성(Device Health), 네트워크 Context, 데이터 민감도 등을 종합 평가한다.  
- **역할:** “허용/거부/추가 인증” 결정을 내려 Policy Administrator로 전달.  
- **특징:** AI or Rule Base 방식으로 정책을 실시간 갱신할 수 있음.

---

#### 2️⃣ **Policy Administrator (정책 관리자)**
- Policy Engine의 판단을 실제 네트워크 수준 명령으로 변환.  
- 방화벽 Rule, API Gateway, VPN Tunnel, SDN Controller 등과 연동.  
- **역할:** 승인된 세션을 설정하거나, 거부 시 연결 차단 명령을 Enforcement Point에 전달.

---

#### 3️⃣ **Policy Enforcement Point (정책 집행 지점)**
- 사용자와 리소스 간 **실제 트래픽 흐름을 제어하는 물리적·논리적 지점.**  
- 네트워크 게이트웨이, 프록시, IoT Edge Gateway 등이 여기에 해당.  
- **역할:** 패킷 수준에서 정책을 적용하며, 세션 중에도 지속적으로 상태를 모니터링한다.  

---

### 📌 프레임워크 데이터 흐름 (Simplified Flow)

1. 사용자·기기·서비스가 리소스 접근을 요청  
2. Policy Engine이 신원, 기기상태, Context 등을 평가  
3. Policy Administrator가 접근 제어 명령을 생성  
4. Policy Enforcement Point가 해당 명령을 실행  
5. 세션 중에도 Policy Engine이 상태 변화를 감시하며 필요 시 정책 재평가  

> 📈 이 구조는 “한 번 인증되면 끝”이 아니라  
> **“세션 중 지속적으로 신뢰를 재검증(Continuous Validation)”** 하는 점이 핵심이다.

---

### 📌 핵심 인사이트
- Zero Trust는 기술이라기보다 **정책·구조적 사고방식(Architecture Mindset)** 이다.  
- “정책 결정(P E) → 정책 전달(P A) → 정책 집행(P E P)” 이라는 세 단계 구조가  
  **보안 SI 프로젝트에서 실제 구현의 기본 Reference Model** 로 사용될 수 있다.  
- 이는 전통적 VPN / ACL 중심 통제에서 벗어나,  
  **클라우드·IoT·엔드포인트가 모두 참여하는 정책 기반 보안**으로 발전하는 과정이다.

---

### 🧠 개인적 해석
- Zero Trust Framework을 SOC 관점에서 보면 “자동 정책 갱신이 가능한 SIEM 모듈”처럼 작동한다.  
- SI 관점에서는 **“설계 표준화의 출발점”** 이다 — 향후 SOAR 또는 SDN 보안 설계와 연계될 가능성이 높다.  
- “보안은 정책이 코드처럼 동작해야 한다(Policy as Code)” 는 철학을 잘 보여준다.

## ✅ Day 3 – Proposed Architecture Design

---

### 📌 제안된 구조의 개요 (Overview)
이 논문은 기존 Zero Trust 원칙을 **Cloud–IoT 통합 환경**에 맞게 구체화한 **3계층 보안 아키텍처**를 제시한다.  
구조는 **Device Layer → Network Layer → Cloud Layer**로 구성되며, 각 계층이 독립적으로 검증·정책 적용을 수행한다.

> 🔹 목표:  
> **“신뢰 없는 연결 환경에서도, 데이터·기기·사용자의 관계를 지속적으로 검증하며 보호”**

---

### 🏗️ 1️⃣ Device Security Layer (IoT 단)
- **역할:**  
  IoT 디바이스의 신원(Identity)과 무결성(Integrity)을 검증하고, 안전한 통신 초기화를 담당한다.  
- **핵심 기능:**  
  - 디바이스 ID·인증서 기반 인증 (Device Certificate or Token)  
  - Firmware Integrity Check (펌웨어 변조 여부 확인)  
  - 데이터 암호화 및 서명 기반 전송  
- **Zero Trust 적용:**  
  모든 IoT 디바이스는 통신 전마다 **Policy Engine에 신뢰도 평가**를 요청해야 한다.

---

### 🌐 2️⃣ Network Security Layer (Micro-Segmentation 단)
- **역할:**  
  네트워크 트래픽을 세분화하여, 침입 또는 이상행동이 한 영역에서 다른 영역으로 확산되지 않도록 차단한다.  
- **핵심 구성요소:**  
  - **SDN Controller:** 네트워크 경로를 소프트웨어적으로 제어해 보안 정책을 동적으로 적용.  
  - **Network Monitor:** 패킷 헤더, 세션 메타데이터를 분석해 비정상 행동 탐지.  
  - **Policy Enforcement Point(PEP):** 패킷 단위로 접근 허용/차단을 집행.  
- **Zero Trust 적용:**  
  동일 네트워크 내에서도 기기 간 통신은 기본 거부(Deny by default) 정책을 따르며,  
  Policy Engine이 허용한 세션만 임시적으로 열린다.

---

### ☁️ 3️⃣ Cloud Access & Policy Layer (정책 및 관리 단)
- **역할:**  
  중앙의 Policy Engine이 클라우드 리소스와 IoT 디바이스 전체를 관리하며, 정책 생성 및 갱신을 수행한다.  
- **핵심 기능:**  
  - Policy Engine → 상황 인식(Context-Aware) 정책 평가  
  - Policy Administrator → 네트워크 / API 레벨 명령 배포  
  - Policy Repository → 정책 버전 관리 및 감사 로그  
- **Zero Trust 적용:**  
  각 요청에 대한 “신원 + 위치 + 행동 이력”을 조합해 **Adaptive Access Control** 결정.

---

### 🔁 데이터 흐름 (End-to-End Process)

1. **IoT Device**가 클라우드 리소스 접근 요청  
2. **Policy Engine**이 신원(ID) + 무결성 + Context 평가  
3. **Policy Administrator**가 승인 세션 규칙을 생성  
4. **PEP**이 트래픽을 허용하고, 패킷 단위 모니터링 시작  
5. 세션 중 이상 징후 발견 시 Policy Engine이 재검증 → 세션 강제 종료 또는 추가 인증 요구  

> 💡 **핵심:**  
> “한 번 인증되면 끝”이 아닌, **세션 중 상시 검증 및 정책 자동 갱신**이 이뤄진다.

---

### 📊 아키텍처 요약 표

| 계층 | 주요 구성요소 | 핵심 기능 | Zero Trust 적용 포인트 |
|------|----------------|------------|-------------------------|
| Device Layer | IoT Device, Firmware Checker | 디바이스 무결성 검증 | 인증 전 Policy Engine 평가 |
| Network Layer | SDN, PEP, Network Monitor | 트래픽 세분화, 패킷 검사 | Micro-Segmentation + Session Control |
| Cloud Layer | Policy Engine, Administrator | 정책 생성 및 배포 | Context-Aware 정책 갱신 |

---

### 📌 핵심 인사이트
- Zero Trust 구조의 핵심은 “중앙 통제형 정책 + 분산 검증형 보안”의 결합이다.  
- 기존 SOC 탐지형 AI가 이벤트 분석에 집중했다면, ZTA는 **시스템 자체의 방어 설계**를 다룬다.  
- Cloud–IoT 통합 구조는 **“보안 경계를 없애고, 정책을 경계로 대체하는 시도”** 이다.  

---

### 🧠 개인적 해석
- 이 구조는 “설계형 보안(Preventive Architecture)”의 전형적인 모델이다.  
- SDN 및 Policy Engine 을 중심으로 하는 보안 자동화는 향후 SOAR 및 Zero Trust Network Access(ZTNA)로 직결될 것이다.  
- Day 4에서 이 구조의 **성능 및 보안 효과 검증 (Experiments & Evaluation)** 을 분석할 예정이다.
