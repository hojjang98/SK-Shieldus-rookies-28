## 데이터베이스 + 로그인 페이지

```python
# 데이터 베이스 생성

import sqlite3

conn = sqlite3.connect('vulnerable_webapp/users.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    username TEXT,
    password TEXT
)
''')

cursor.execute("INSERT INTO users (username, password) VALUES ('admin', 'password123')")
cursor.execute("INSERT INTO users (username, password) VALUES ('user1', 'mypass')")

conn.commit()
conn.close()
```

```python
# 로그인 HTML 파일 생성

login_html = '''
<!DOCTYPE html>
<html>
<head>
    <title>로그인</title>
</head>
<body>
    <h2>로그인</h2>
    <form method="POST" action="/login">
        <input type="text" name="username" placeholder="아이디">
        <input type="password" name="password" placeholder="비밀번호">
        <button type="submit">로그인</button>
    </form>
    <p style="color: red;">{{ message }}</p>
</body>
</html>
'''

with open('vulnerable_webapp/templates/login.html', 'w', encoding='utf-8') as f:
    f.write(login_html)
```

```python
#  Flask 앱 업데이트 (취약한 로그인)

from flask import Flask, render_template, request
import sqlite3

app = Flask(__name__, template_folder='vulnerable_webapp/templates')

@app.route('/')
def home():
    return render_template('login.html')

@app.route('/login', methods=['POST'])
def login():
    username = request.form['username']
    password = request.form['password']
    
    # 취약한 SQL 쿼리 (SQL Injection 가능)
    conn = sqlite3.connect('vulnerable_webapp/users.db')
    cursor = conn.cursor()
    query = f"SELECT * FROM users WHERE username='{username}' AND password='{password}'"
    cursor.execute(query)
    user = cursor.fetchone()
    conn.close()
    
    if user:
        return f"<h2>로그인 성공!</h2><p>환영합니다, {username}님</p>"
    else:
        return render_template('login.html', message="로그인 실패")
```

## 게시판 기능 추가

```python
# 게시판 테이블 생성

conn = sqlite3.connect('vulnerable_webapp/users.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS posts (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    content TEXT
)
''')

conn.commit()
conn.close()
```

```python
# 게시판 HTML 파일 생성

board_html = '''
<!DOCTYPE html>
<html>
<head>
    <title>게시판</title>
</head>
<body>
    <h2>게시판</h2>
    
    <h3>글 작성</h3>
    <form method="POST" action="/post">
        <input type="text" name="title" placeholder="제목" required><br>
        <textarea name="content" placeholder="내용" required></textarea><br>
        <button type="submit">작성</button>
    </form>
    
    <hr>
    
    <h3>게시글 목록</h3>
    {% for post in posts %}
        <div>
            <h4>{{ post[1] }}</h4>
            <p>{{ post[2]|safe }}</p>
            <hr>
        </div>
    {% endfor %}
</body>
</html>
'''

with open('vulnerable_webapp/templates/board.html', 'w', encoding='utf-8') as f:
    f.write(board_html)
```

```python
# Flask 앱에 게시판 라우트 추가

from flask import Flask, render_template, request, redirect
import sqlite3

app = Flask(__name__, template_folder='vulnerable_webapp/templates')

@app.route('/')
def home():
    return render_template('login.html')

@app.route('/login', methods=['POST'])
def login():
    username = request.form['username']
    password = request.form['password']
    
    conn = sqlite3.connect('vulnerable_webapp/users.db')
    cursor = conn.cursor()
    query = f"SELECT * FROM users WHERE username='{username}' AND password='{password}'"
    cursor.execute(query)
    user = cursor.fetchone()
    conn.close()
    
    if user:
        return redirect('/board')
    else:
        return render_template('login.html', message="로그인 실패")

@app.route('/board')
def board():
    conn = sqlite3.connect('vulnerable_webapp/users.db')
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM posts")
    posts = cursor.fetchall()
    conn.close()
    return render_template('board.html', posts=posts)

@app.route('/post', methods=['POST'])
def post():
    title = request.form['title']
    content = request.form['content']
    
    conn = sqlite3.connect('vulnerable_webapp/users.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO posts (title, content) VALUES (?, ?)", (title, content))
    conn.commit()
    conn.close()
    
    return redirect('/board')
```

```python
import threading

def run_app():
    app.run(debug=False, use_reloader=False)

thread = threading.Thread(target=run_app)
thread.daemon = True
thread.start()
```
